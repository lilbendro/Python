```python
#######################################################################
# Encuentas versión 1
#######################################################################
# Requerimientos funcionales:
#
#   -> Preguntar Edad (validar)
#   -> Preguntar puntuación de 0 a 10 (validar)
#   -> Preguntar si es el último encuestado
#   -> Con el último encuestado mostrar el número de encuentas 
#      y la puntuación media
#######################################################################

opinion = []
ultimo = "no"

while (ultimo == "no"):
    try:
        valor = input("Satisfación de 1 a 10: ")

        if (valor.lower() == "fin"):
            break
        elif (valor.isalpha()):
            print(f"El valor {valor} no es valido.")
            print("")
        elif (int(valor) < 1 or int(valor) > 10):
            print(f"El valor {valor} no es valido.")
            print("")
        else:
            opinion.append(int(valor))
    except Exception as err:
        print(f"Error: {err}")
        print("")
    ultimo = input("¿Eres el último").lower() #Todo el input se convierte a minuscula
    
#falta hacer la edad

print("")
print(f">> Número de Encuestados: {len(opinion)}")
print(f">> Satisfación Media: {sum(opinion) / len(opinion)}")
    


    
```

```python
#######################################################################
# Hucha versión 1
#######################################################################
# Requerimientos funcionales:
#
#   -> Preguntar la acción: Insertar dinero // Romper la hucha
#   -> Romper hucha mostramos el importe total
#   -> Insertar: 50, 100, 200, 500 o 1000
#   -> Después de insertar mostra el importe total y el número 
#      de monedas de cada tipo
#######################################################################

from os import system

hucha = {"50": 0, "100": 0, "200": 0, "500": 0, "1000": 0}
total = 0

try:
    while (True):
        accion = input(
            "R para romper hucha, I para insertar monedas: ").upper()

        if (accion == "I"):
            while (True):
                moneda = input(
                    "Valor de la moneda (50, 100, 200, 500 o 1000): ")
                if (moneda in hucha.keys()): #Evalua si la moneda está dentro de los valores de las claves
                    system("cls") #Borrado de pantalla de Windows con una sentencia del sistema

                    hucha[moneda] += 1 #Suma una moneda a esa clave, al pasar antes por un if se ha validado que el valor moneda está dentro de las claves
                    total = 0
                    
                    for key in hucha:
                        print(f"{hucha[key]} monedas de {key}")
                        total += int(key) * hucha[key] #Múltiplica el número de moendas por su valor

                    print(f"\nImporte total: {total:1.2f}\n")

                    break
                else:
                    print(f"La moneda {moneda} no es de un valor valido.")
        elif (accion == "R"):
            system("cls")

            print(f"\nImporte total: {total:1.2f}\n")

            break
except Exception as err:
    print("No se puede mostrar información sobre la hucha.")
    print(err)




```

```python
#######################################################################
# Requerimientos funcionales:
#
#   -> Calcular la letra de DNI
#   -> Formula: dni % 23
#######################################################################

letras = ['T', 'R', 'W', 'A', 'G', 'M', 'Y', 'F', 'P', 'D', 'X', 'B', 'N', 'J', 'Z', 'S', 'Q', 'V', 'H', 'L', 'C', 'K', 'E', 'T'] 
#Construimos una colección para las letras que sigue el orden de 0 a 14 de las letras que se asocian al resto

try:
    dni = int(input("Escribe tu número de DNI: "))
    resto = dni % 23 #El resto funcionará como un índice de la colección

    print(f"La letra del DNI {dni} es {letras[resto]}") 

except Exception as err:
    print(f"Error: {err}")


```
